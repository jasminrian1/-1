Consider the following hierarchy that represents different types of neurons and their roles in the nervous system :

```python
class GENERALNEURON:
  def __init__(self,firing_rate):
      self.firing_rate=firing_rate
  def activate(self,stimulus):
   pass
                        

class SENSORYNEURON(GENERALNEURON):
   def __init__(self,firing_rate,receptor_type):
      super().__init__(firing_rate)
      self.receptor_type=receptor_type
   def Sense_Stimulus(self,stimulus):
      pass

class MOTORNNEURON(GENERALNEURON):
   def __init__(self , firing_rate , target_muscle):
      super().__init__(firing_rate)
      self.target_muscle=target_muscle
   def ControlMuscle(self, stimulus):
      pass
   
class Photoreceptor (SENSORYNEURON):
   def __init__(self, firing_rate):
      super().__init__(firing_rate , receptor_type="light")
   def LightDetection(self , stimulus):
      pass
   
class Mechanoreceptor (SENSORYNEURON):
   def __init__(self, firing_rate):   
      super().__init__(firing_rate , receptor_type="pressure")
   def PressureDetection(self , stimulus):
      pass   

class AlphaMotorNeuron (MOTORNNEURON):
   def __init__(self , firing_rate , target_muscle):
      super().__init__(firing_rate)
      self.target_muscle=target_muscle
   def SkeletalMuscleControl(self, stimulus):
      pass
   
class GammaMotorNeuron (MOTORNNEURON):
   def __init__(self , firing_rate , target_muscle):
      super().__init__(firing_rate)
      self.target_muscle=target_muscle
   def MuscleSpindleControl(self, stimulus):
      pass
```

Q1 (20pts): Write a function that samples a sinusoidal signal while preserving all maximum and minimum points.
Input to function: a NumPy array representing a sinusoidal signal with a frequency of 5Hz, sampled every 0.01 seconds.
Output: a new numpy array representing the same signal. The length of the new array should be approximately 20% of the original one.

לפי מה שהבנתי , נדרשת כתיבת פונקציה שחוזרת על עצמה, תוך צמצום מספר הנקודות ושמירה על נקודות המקסימום והמינימום.

הגדרתי תדר של 5 הרץ, כלומר חמש מחזורי גל בשנייה, ולקחתי דגימות כל 0.01 שניות. לאחר מכן חיפשתי את נקודות המקסימום והמינימום באמצעות השוואה של כל נקודה עם שכנותיה: אם הנקודה גדולה משתיהן, היא מקסימום; ואם היא קטנה משתיהן, היא מינימום. לאחר מכן צמצמתי את מספר הנקודות על ידי בחירת נקודה אחת מתוך כל חמש נקודות, בהתאם לדרישה לצמצם את מספר הנקודות בכ-20%. וידאתי שנקודות המקסימום והמינימום נשמרו ושנכללו בנקודות שנבחרו. לבסוף, השתמשתי בספריית Matplotlib ליצירת גרף שמציג את הפתרון.

```python
#q1

import numpy as np
import matplotlib.pyplot as plt

# הפונקציות שהשתמשתי בהן להשוות בין הנקודות ולבחור את המינמום והמקסימום ואז להחזיר אותן הן minima / maxima
# minima and maxima function מתייחסות לפונקציות או חישובים שמטרתם למצוא את נקודות המקסימום והמינימום של סדרת נתונים או פונקציה.
# minima and maxima function   כתבתי לו מה זהchat gpt את ההסבר שכתבתי למעלה לקחתי מ 
import numpy as np

# יצירת סדרת נתונים לדוגמה
data = [1, 3, 7, 1, 2, 6, 0, 1]

# מציאת אינדקסי מקסימום ומינימום
maxima = [i for i in range(1, len(data)-1) if data[i] > data[i-1] and data[i] > data[i+1]]
minima = [i for i in range(1, len(data)-1) if data[i] < data[i-1] and data[i] < data[i+1]]

print("Maxima indices:", maxima)
print("Minima indices:", minima)





def find_maxima_minima():
    maxima = [i for i in range(1, len(np.sin(2 * np.pi * 5 * time)) - 1) if 
              np.sin(2 * np.pi * 5 * time)[i] > np.sin(2 * np.pi * 5 * time)[i - 1] and 
              np.sin(2 * np.pi * 5 * time)[i] > np.sin(2 * np.pi * 5 * time)[i + 1]]
    minima = [i for i in range(1, len(np.sin(2 * np.pi * 5 * time)) - 1) if 
              np.sin(2 * np.pi * 5 * time)[i] < np.sin(2 * np.pi * 5 * time)[i - 1] and 
              np.sin(2 * np.pi * 5 * time)[i] < np.sin(2 * np.pi * 5 * time)[i + 1]]
    return maxima, minima

#הנה הפונקציה שבה השתמשתי כדי לצמצם את מספר הנקודות תוך שמירה על נקודות המקסימום והמינימום:
def downsample_signal():
    downsampled_indices = list(range(0, len(np.sin(2 * np.pi * 5 * time)), 5))
    maxima, minima = find_maxima_minima()

#finalיצרתי רשימה חדשה: התחלתי עם העתקה של הרשימה שיצרתי קודם לרשימה חדשה בשם 
 
    final = downsampled_indices.copy()

#combined_induces אחרי שמצאתי את נקודות המאקסימום והמינימום ,שמתי את כולן ברשימה חדשה בשם 
 # הרשימה הזו תכיל את כל האינדיקסים של המאקסימום והמינימום 
 # אחרי זה השתמשתי בלולאה בשביל לבדוק אם מה שעשיתי עד עכשיו בסדר , כאילו בעזרת זה בדקתי את החפיפה בין שתי הרשימות שעשיתי ואם משהו היה חסר אז הוספתי אותו לרשימה הקודמת ( התקוונתי לאינדיקסים)
 # ולאחר מכן רציתי לוודא שהאינדיקסים יסודרו מהקטן לגדול   
    combined_indices = maxima + minima
    i = 0
    while i < len(combined_indices):
        if combined_indices[i] not in final:
            final.append(combined_indices[i])
        i += 1

#עשיתי פקודה שתחזיר לי את הרשימה הסופית של הנקודות שאני רוצה עם המחזוריות שבחרתי שהי 5 
# נסיתי להסביר את מה שעלה לי בראש בצורה הכי ברורה :)     
    final.sort()
    downsampled_signal = np.sin(2 * np.pi * 5 * time)[final]
    return final, downsampled_signal


time = np.arange(0, 1, 0.01)
indices, downsampled_signal = downsample_signal()
# כאן התחלתי לעבוד על הציור של התוצאות כמו גודל ת צבע והתוצאות ואת הצירים של הגרם את השמות שלהם וכולי...
plt.figure(figsize=(12, 7))
plt.fill_between(time, np.sin(2 * np.pi * 5 * time), color="lightblue")
plt.plot(time, np.sin(2 * np.pi * 5 * time), label="Original Signal", color="blue", linewidth=2)
plt.scatter(time[indices], downsampled_signal, color="red", s=50, label="Downsampled Points", zorder=5)
plt.plot(time[indices], downsampled_signal, color="orange", linestyle="--", linewidth=2, label="Downsampled Signal")
plt.xlabel("Time (seconds)", fontsize=14)
plt.ylabel("Amplitude", fontsize=14)
plt.title("Signal Downsampling", fontsize=16)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.5)
plt.show()
```

Q2 (65pts): For this question you will use the dataset titled “laptop-price – dataset.csv”.
Import the libraries mentioned above and import the dataset from your filesystem into the code.

בשאלה הזו שמתי כל חלק בפלוק כדי שכל קוד יהיה לו פלט מתחתיו ,ראיתי שככה יותר ברור 

```python
#q2
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# בהתחלה שמרתי את הקובץ ב dataset 
laptop_price_dataset = pd.read_csv('laptop_price - dataset.csv')      

# yיהיה האינדיקס והמחיר יהיה ב x בו התחלתי בלעשות את הגרף ,עשיתי שציר ה

plt.figure(figsize=(14, 8))
plt.scatter(
    laptop_price_dataset.index, 
    laptop_price_dataset['Price (Euro)'], 
    color='cyan', 
    marker='s', 
    s=75, 
    label='Laptop Prices'
)
# בו עשיתי את התיאור של הפלוט 
for index, price in zip(laptop_price_dataset.index, laptop_price_dataset['Price (Euro)']):
    plt.plot([index, index], [0, price], linestyle='--', color='gray', linewidth=0.8, alpha=0.7)
plt.title('Laptop Prices Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Laptop Index', fontsize=13, color='darkred')
plt.ylabel('Price in Euros', fontsize=13, color='darkred')
plt.grid(visible=True, linestyle=':', linewidth=0.7, alpha=0.6)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# בו ביקשו בשאלה לדעת איזה חברה יש לה ממוצע המחירים הכי גדול
# הפקודה הזו תחשב לי את הממוצע הזה לכל חברה לבד ואס תסדר אותן
Avarg = laptop_price_dataset.groupby(['Company']).agg({'Price (Euro)':
                                                      'mean'}).sort_values(by='Price (Euro)', 
                                                                                          ascending=False)['Price (Euro)']
companies_and_prices = list(zip(Avarg.index, Avarg.values))
index = 0
while index < len(companies_and_prices):
    company, price = companies_and_prices[index]
    print(f"{company}: {price:.3f} Euros")
    index += 1
exp_company, exp_company_price = companies_and_prices[0]
# אחרי שחישבתי את הממוצע של המחירים וסידרתי אותו מהגדול לקטן ביקשתי להדפיס את הרשימה בשמות של החברות עם הממוצע של כל אחת מהן
print("\nCompany with the Most Expensive Laptops on Average")
print(f"{exp_company} - with Avg price: {exp_company_price:.3f}\n")
```

```python
#  windowsלמשל כשם אחד שהו windows ואז שמה את כל השמות שבהן operator system  השתמשתי ב פונקצייה שלוקחת את השם של ה 
mac = {'Mac OS X'}
windows = {'Windows 7', 'Windows 10 S', 'Windows 10'}
def uni_v2(os_name):
    if os_name in mac:
        return 'macOS'
    elif os_name in windows:
        return 'Windows'
    return os_name
# operating systemבו ביקשתי שישתמש בעמודה של ה 
laptop_price_dataset['OpSys'] = [uni_v2(os_name) for os_name in laptop_price_dataset['OpSys']]
print("Operating Systems:")
print(laptop_price_dataset['OpSys'].unique())
laptop_price_dataset.to_csv('laptop_price - dataset.csv', index=False)
```

```python

grouped = laptop_price_dataset.groupby('OpSys')
for currently, group in grouped:
    prices = group['Price (Euro)']
    plt.figure(figsize=(12, 6))
    plt.hist(
        prices, bins=15, color='teal', alpha=0.85, edgecolor='darkblue')
    plt.title(f"Price Distribution of {currently}", fontsize=16, fontweight='bold')
    plt.xlabel("Price in Euros", fontsize=14, fontstyle='italic', color='darkred')
    plt.ylabel("Laptop Count", fontsize=14, fontstyle='italic', color='darkred')
    plt.grid(axis='y', linewidth=0.7, alpha=0.6)
    plt.xticks(fontsize=12, rotation=45, color='darkblue')
    plt.yticks(fontsize=12, color='darkblue')
    plt.tight_layout()
    plt.show()
```

```python
# בו חישבתי את הממוצע של המחירים ואז עשיתי גרפים שמציגים את הממוצע בהשוואה לדודל הראם
ram_avg_prices = laptop_price_dataset.pivot_table(values='Price (Euro)', index='RAM (GB)', aggfunc='mean').reset_index()
ram_sizes = ram_avg_prices['RAM (GB)']
avg_prices = ram_avg_prices['Price (Euro)']
fig, plot = plt.subplots(figsize=(10, 6))
bars = plot.bar(ram_sizes, avg_prices, color='skyblue', edgecolor='navy', alpha=0.8, width=0.9, label='Average Price')
plot.step(ram_sizes, avg_prices, where='mid', color='brown', linewidth=2.5)
plot.set_title("Average Laptop Price by RAM Size", fontsize=16, fontweight='bold')
plot.set_xlabel("RAM Size", fontsize=14, labelpad=10)
plot.set_ylabel("Average Price", fontsize=14, labelpad=10)
plot.set_xticks(ram_sizes)
plot.set_xticklabels(ram_sizes, fontsize=12, rotation=30)
plot.tick_params(axis='y', labelsize=12, colors='darkred')
plot.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)
plot.legend(loc='best', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# חשבתי ככה : אם יש לי בעמדה + אז יש לי יותר מסוג אחד ולחן איחדתי בינם ועשיתי אותן כסוג אחד ואם לא היה לי + אז החזרתי את המילה עצמה 
def extract_storage_types(memory):
    if "+" not in memory:
        val = memory.split(maxsplit=1)[1]
    else:
        val="  ".join(set(part.split(maxsplit=1)[1] for part in memory.split(" + ")))
    return val
laptop_price_dataset["Storage type"] = laptop_price_dataset["Memory"].apply(extract_storage_types)
laptop_price_dataset.to_csv('laptop_price - dataset.csv', index=False)
```
